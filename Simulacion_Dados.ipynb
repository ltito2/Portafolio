```python
from numpy import random
import matplotlib.pyplot as pp

simulacion = [100,1000,10000,100000]
pp.figure(figsize=(10,10))
for i in range(4):
#1.Realizar la generación de un número del 1-6 (dado)
  dado1 = random.randint(1,7,simulacion[i])
  dado2 = random.randint(1,7,simulacion[i])
#2.Sumar el valor de los dos dados (espacio valores 2 - 12)
  suma = dado1+dado2
  frecuencia = []
#3.Obtener la frecuencia (reperición de la sumatoria)
#4.Generar simulaciones de 100 veces de lanzamiento, 1000, 10000, 100000
  for j in range(simulacion[i]):
    frecuencia.append(suma[j])
#5.Finamente generar un histograma de frecuencia
  if i == 0:
    for j in range(2,13):
      x1.append(j)
      y1.append(frecuencia.count(j))
    pp.subplot(221)
    pp.bar(x1, y1)
    pp.grid(color='b', linestyle='-', linewidth=1)
    pp.xlabel('Sumas')
    pp.ylabel('Veces')
    pp.xlim(2,12)
    pp.title('Histograma de 100')
  elif i == 1:
    for j in range(2,13):
      x2.append(j)
      y2.append(frecuencia.count(j))
    pp.subplot(222)
    pp.bar(x2, y2)
    pp.grid(color='g', linestyle='-', linewidth=1)
    pp.xlabel('Sumas')
    pp.ylabel('Veces')
    pp.xlim(2,12)
    pp.title('Histograma de 1000')
  elif i == 2:
    for j in range(2,13):
      x3.append(j)
      y3.append(frecuencia.count(j))
    pp.subplot(223)
    pp.bar(x3, y3)
    pp.grid(color='c', linestyle='-', linewidth=1)
    pp.xlabel('Sumas')
    pp.ylabel('Veces')
    pp.xlim(2,12)
    pp.title('Histograma de 10000')
  else:
    for j in range(2,13):
      x4.append(j)
      y4.append(frecuencia.count(j))
    pp.subplot(224)
    pp.bar(x4, y4)
    pp.grid(color='r', linestyle='-', linewidth=1)
    pp.xlabel('Sumas')
    pp.ylabel('Veces')
    pp.xlim(2,12)
    pp.title('Histograma de 100000')
```
